# **Project name :** Mobile price classification using Machine learning.

 Many people regularly buy new mobile phones as they come with exciting new features, making them the most popular electronic devices that are sold in large numbers every day. This makes it challenging for someone starting a mobile phone business to determine the appropriate price for their products.

 **Mobile price classification** using machine learning can be achieved using the Mobile Price Classification dataset from Kaggle. The dataset contains several features such as battery power, RAM, internal memory, etc., which can be used to classify the price range of mobile phones as 0 (low cost), 1 (medium cost), 2 (high cost), or 3 (very high cost).

# Preprocess the data

- This dataset only has numbers as features, so we don't need to make any changes to it. However, before we can train a model, we should standardize or normalize the data and split it into separate training and testing sets.

```python
# x_train containing all the rows of the df_train except the last column (the target variable price_range)
x=df_train.iloc[:, :-1] 
# y_train containing the target variable price_range of the df_train 
y=df_train.iloc[:, -1] 
# Split the data into training and testing sets
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=70) 
# creat an instance of the StandardScaler() class which is used for standardization of the data.
sc=StandardScaler()
# comput the mean and standard deviation of each feature and then scales and centers the data based on these values.
x_train=sc.fit_transform(x_train)
# The testing data is transformed in the same way as the training data, using the same parameters, so that the model can generalize well to unseen data.
x_test=sc.fit_transform(x_test)
```

# Train the model
- After preprocessing the data, we can move forward to train our machine learning model. To begin with, we will create the following baseline models: Random Forest Classifier, Support Vector Classifier and Logistic Regression.

- Random Forest Classifier

```python
model_clf=RandomForestClassifier() # Initializing the Random Forest Classifier
model_clf.fit(x_train, y_train) # Training the model
```

- SVM

```python
model_SVM= svm.SVC()
model_SVM.fit(x_train, y_train)
```

- Logistic Regression

```python
model_LR= LogisticRegression()
model_LR.fit(x_train, y_train)
```

# Test the performance of the model on the test set.

- Random Forest Classifier

```python
y_pred_clf=model_clf.predict(x_test) # Predicting the price range of mobile phones in the test set
Accuracy=accuracy_score(y_test, y_pred_clf) # Testing the model using the test set
print(f'Accuracy {Accuracy}')
```

```python
#A report with precision, recall, F1-score, and support for each class.
report=classification_report(y_test,y_pred_clf)
print(f'Report: {report}')
```

```python
# assuming y_pred_clf and y_test are the predicted and true labels respectively
cm = confusion_matrix(y_test, y_pred_clf)
plt.title('Confusion Matrix for Random Forest Classifier')
# Plot the confusion matrix as a heatmap
sns.heatmap(cm, annot=True, cmap='Reds', fmt='g')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

- SVM

```python
y_pred_svm=model_SVM.predict(x_test) 
Accuracy=accuracy_score(y_test, y_pred_svm) 
print(f'Accuracy {Accuracy}')
```

```python
report=classification_report(y_test,y_pred_svm)
print(report)
```

```python
cm = confusion_matrix(y_test, y_pred_svm)
plt.title('Confusion Matrix for SVM')
sns.heatmap(cm, annot=True, cmap='Reds', fmt='g')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

- Logistic Regression

```python
y_pred_lr=model_LR.predict(x_test) 
Accuracy=accuracy_score(y_test, y_pred_lr) 
print(f'Accuracy {Accuracy}')
```

```python
report=classification_report(y_test,y_pred_lr)
print(report)
```
